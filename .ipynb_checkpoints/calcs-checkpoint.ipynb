```python
from  functools import partial
from scipy import integrate
from scipy.optimize import broyden1, broyden2, newton_krylov, fixed_point, fsolve, minimize, least_squares

kappa = 0.5

def mm(x):
    return np.minimum(np.maximum(x,0),1)

def acop(x,y,kappa):
    return ( 1 + kappa * ((1+x)*(1+y)-3) + (kappa**2)*(1-x)*(1-y)) / ((1-kappa*(1-x)*(1-y))**3)

def func_b(k,m):
    out, err = integrate.dblquad(lambda x,y: 1, 0,1, lambda x: mm(m + x * k), 1)
    return out - 1/2

def det_m(k):
    return broyden2(partial(func_b,k),  [0.5], f_tol=1e-14)[0]

def func_x(k):
    m = det_m(k)
    ans1, err1 = integrate.dblquad(lambda x,y: x*acop(x,y,kappa), 0 , 1, lambda x: m + x * k,  1)
    ans0, err0 = integrate.dblquad(lambda x,y: y*acop(x,y,kappa), 0,1 , 0, lambda x: m + x * k )
    return (ans0/ans1)**3

least_squares(lambda k: func_x(k)**2-k,  [0.3], bounds=([0, 1]))
#scipy.optimize.fixed_point(func_x(k)**3, [0.2] , maxiter=500)
subscript = str.maketrans("0123456789", "₀₁₂₃₄₅₆₇₈₉")
vecfunc = np.vectorize(func_x)
d = np.arange(0.001, 1.04, 0.03)
fig, ax = plt.subplots( figsize=(10, 10))
ax.plot(d,vecfunc(d),label='Separating equilibrium')
vecfunc2 = np.vectorize(lambda x: x)
ax.plot(d,vecfunc2(d), label='Pooling equilibrium')
ax.set_ylim([0, 1.0])
ax.set_xlim([0, 1.0])
ax.legend()
plt.show()
```
